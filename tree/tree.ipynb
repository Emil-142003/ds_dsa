```python
root=Tree(5)
root.right=Tree(8)
root.right.right=Tree(100)
root.left=Tree(4)
root.left.left=Tree(3)
```

```python
def inorder(root):
    if root:
      inorder(root.left)
      print(root.data)
      inorder(root.right)
inorder(root)
```

```python
def preorder(root):
    if root:
        print(root.data)
        preorder(root.left)
        preorder(root.right)
preorder(root)
```

```python
def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.data)
postorder(root)
```

```python
class Tree:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
        
```

```python
root=Tree(5)
root.right=Tree(8)
root.right.right=Tree(100)
root.left=Tree(4)
root.left.left=Tree(3)
```

```python
num = int(input("Enter the element to be searched: "))
def find_level(root, num, level):
    if root is None:
        return -1
    if root.data == num:
        return level
    if num < root.data:
        return find_level(root.left, num, level + 1)
    else:
        return find_level(root.right, num, level + 1)
level = find_level(root, num, 0)
if level == -1:
    print("Element not found")
else:
    print("Height of", num, "is", level)
```

```python
class Tree:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
def insert(root, data):
    if root is None:
        return Tree(data)
    if data < root.data:
        root.left = insert(root.left, data)
    else:
        root.right = insert(root.right, data)
    return root
def inorder(root):
    if root:
        inorder(root.left)
        print(root.data)
        inorder(root.right)
def preorder(root):
    if  root:
        print(root.data)
        preorder(root.right)
        preorder(root.left)
def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.data)
n = int(input("Enter the number of elements: "))
root = None
print("Enter the elements:")
for i in range(n):
    data = int(input())
    root = insert(root, data)
print("\nInorder Traversal of BST:")
inorder(root)
print("\nPreorder Traversal of BST:")
preorder(root)
print("\nPostorder Traversal of BST:")
postorder(root)
```
