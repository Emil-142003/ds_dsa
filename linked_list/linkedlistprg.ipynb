```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
n1 = Node(10)
n2 = Node(20)
n3 = Node(30)
n4 = Node(40)
n1.next = n2
n2.next = n3
n3.next = n4
head = n1
```

```python
cur = head
total = 0
while cur:
    total += cur.data
    cur = cur.next
print("Sum =", total)
```

```python
cur = head
maximum = head.data
while cur:
    if cur.data > maximum:
        maximum = cur.data
    cur = cur.next
print("Maximum =", maximum)
```

```python
prev = None
cur = head
while cur:
    nxt = cur.next
    cur.next = prev
    prev = cur
    cur = nxt
head = prev
```

```python
cur = head
while cur:
    print(cur.data, end=" ")
    cur = cur.next
```

```python
class Student:
    def __init__(self,id,name,result):
        self.id=id
        self.name=name
        self.result=result
        self.next=None
    def std_info(self):
        return(f"id:{self.id} name:{self.name} result:{self.result}")
```

```python
n=int(input("Enter how many students:"))
students=[]
```

```python
for i in range(n):
    id=int(input("enter student id"))
    name=input("enter student name")
    result=input("enter student result")
    s1=Student(id,name,result)
    students.append(s1)
```

```python
for s1 in students:
    s1.std_info()
```
