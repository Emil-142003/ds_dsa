```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
n1 = Node(10)
n2 = Node(20)
n3 = Node(40)
n4 = Node(30)
n1.next = n2
n2.next = n3
n3.next = n4
n4.next = n1  
cur = n1
largest = -1
second_largest = -1
start = n1
while True:
    if cur.data > largest:
        second_largest = largest
        largest = cur.data
    elif cur.data > second_largest and cur.data != largest:
        second_largest = cur.data
    cur = cur.next
    if cur == start:
        break
print("Largest node:", largest)
print("Second largest node:", second_largest)
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
n1 = Node(10)
n2 = Node(7)
n3 = Node(11)
n4 = Node(15)
n5 = Node(20)
n1.next = n2
n2.next = n3
n3.next = n4
n4.next = n5
head = n1
n = int(input("Enter the node position: "))
if n == 1:
    head = head.next
else:
    cur = head
    count = 1
    while cur.next != None and count < n - 1:
        cur = cur.next
        count += 1
    if cur.next != None:
        cur.next = cur.next.next
print("The node", n, "is deleted")
print("The nodes after deletion:")
cur = head
while cur != None:
    print(cur.data, end=" ")
    cur = cur.next


```
