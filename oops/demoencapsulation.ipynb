```python
class Customer:
    def __init__(self,id,name,address):
        self.id=id
        self.name=name
        self.address=address
    def cust_info(self):
        print(f"Id:{self.id} name:{self.name} address:{self.address}")
```

```python
c=Customer(100,"joe","banglore")
```
